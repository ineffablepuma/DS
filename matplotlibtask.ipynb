```python
# How do you plot a simple line graph using Matplotlib?

import matplotlib.pyplot as plt

x = [1,3,5,7,9,11]
y = [23,75,35,96,56,23]
plt.plot(x,y, label = "Data line", color = "Orange", marker = "o", linestyle = "--",linewidth = "2")
plt.title("This is the title")
plt.xlabel("This is x axis label")
plt.ylabel("This is y axis label")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.plot(x, y, marker='o', color='blue', label='Line Plot')
plt.legend()
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.scatter(x, y, color='red', label='Scatter Plot')
plt.legend()
plt.show()
```

```python
# How do you change the line color and style in a line plot?

# -using color and linestyle.

import matplotlib.pyplot as plt

x = [1,3,5,7,9,11]
y = [83,15,75,53,23,89]
plt.plot(x,y, label = "Data line", color = "Teal", marker = "o", linestyle = ":",linewidth = "2")
plt.title("This is the title")
plt.xlabel("This is x axis label")  #How can you set the x-axis and y-axis labels?
plt.ylabel("This is y axis label")
plt.legend()    #How do you add a legend to a plot? => also there should be label in plt.plot
plt.show()
```

```python
# Which function is used to create histograms?
# plt.hist() is used to create histogram
import matplotlib.pyplot as plt

ages = [22, 25, 29, 30, 32, 35, 36, 40, 42, 45, 50, 52, 55, 60, 61, 65, 70, 72, 75, 80]

plt.hist(ages, bins=8, color='magenta', edgecolor='black')

plt.title("Age Distribution")
plt.xlabel("Age Groups")
plt.ylabel("Frequency")

plt.show()
```

```python
# What does the alpha parameter control?
#controls transparency

import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 8, 10]
y2 = [1, 3, 5, 7, 9]

plt.plot(x, y1, color='blue', label='Opaque Line', alpha=1)

plt.plot(x, y2, color='red', label='Transparent Line', alpha=0.1)

plt.legend()
plt.show()
```

```python
# How do you plot multiple subplots in the same figure?

# by using plt.subplot

import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 8, 10]
y2 = [1, 3, 5, 7, 9]

plt.subplot(2, 1, 1)  # 2 rows, 1 column, first subplot
plt.plot(x, y1, color='blue')
plt.title("Line 1")

plt.subplot(2, 1, 2)  # 2 rows, 1 column, second subplot
plt.plot(x, y2, color='red')
plt.title("Line 2")

plt.tight_layout()  #overlap hunxa yo halena bhane
plt.show()
```

```python
# What is the difference between plt.bar() and plt.barh()?

# plt.bar() gives vertical, default bargraph where category is in x axis and values are in y axis
# plt.barh() gives hotizontal bargraph where category is in y axis and values are in x axis

import matplotlib.pyplot as plt

categories = ['A', 'B', 'C', 'D']
values = [10, 25, 15, 30]

plt.bar(categories, values, color='skyblue')
plt.title("Vertical Bar Chart")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

```python
plt.barh(categories, values, color='lightgreen')
plt.title("Horizontal Bar Chart")
plt.xlabel("Values")
plt.ylabel("Categories")
plt.show()
```

```python
# How do you save a Matplotlib figure as an image?

# plt.savefig() function before calling plt.show().

import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 20, 15, 25, 30]

plt.plot(x, y, color='blue', marker='o')
plt.title("Sample Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

plt.savefig("line_plot.png") 

plt.show()
```
