What is a Histogram?
A histogram divides the range of values into intervals called bins.

Unlike a bar chart (which is for categorical data), a histogram is for continuous or numerical data.

Pie Chart

What it is:
A circle divided into slices, where each slice shows a percentage or fraction of the whole.

When to use:
To show proportions (e.g., "Market share of smartphone companies").

```python
labels = ['Apple', 'Samsung', 'Xiaomi', 'Others']
sizes = [30, 25, 20, 25]
# explode = [0, 0, 0.1, 0.1]
plt.pie(sizes, labels=labels, autopct='%1.2f%%', startangle=90, explode = [0, 0, 0.2, 0.1])
# Explode brings out a part out from the piechart
plt.title("Pie Chart")
plt.axis('equal')  # To make it a circle
plt.show()
```

Stack Plot

What it is:
A line plot where multiple datasets are stacked on top of each other.
It shows how different categories contribute to the total over time.

days → x-axis values.

sleeping, eating, working, playing → multiple datasets stacked on top of each other.

labels → Names for each dataset (used in the legend).

colors → Colors for each stack:

'm' = magenta (Sleep)

'c' = cyan (Eat)

'r' = red (Work)

'k' = black (Play)

Syntax: plt.stackplot(x, y1, y2, y3, ...)

x → values for the horizontal axis

y1, y2, ... → each array represents a layer of the stack (vertical values stacked on top of each other)

Optional: colors=..., labels=...


```python
days = [1, 2, 3, 4, 5]
sleeping = [7, 8, 6, 11, 7]
eating = [2, 3, 4, 3, 2]
working = [7, 8, 7, 2, 2]
playing = [8, 5, 7, 8, 13]

plt.stackplot(days, sleeping, eating, working, playing,
              labels=['Sleep', 'Eat', 'Work', 'Play'],
              colors=['blue','c','r','black'])
plt.legend(loc='lower left')
plt.title('Stack Plot Example')
plt.xlabel('Days')
plt.ylabel('Hours')
plt.xticks(days)  # Set x-ticks to match the days
plt.show()
```

Box Plot (Whisker Plot)

What it is:
Shows the spread of data:

The box shows the middle 50% of data (quartiles).

The line inside the box is the median.

Whiskers show the min and max values.

Dots outside whiskers are outliers.

What Does a Box Plot Show?

Median (Q2): The middle value of the dataset.

Q1 (25th percentile): The value below which 25% of the data lies.

Q3 (75th percentile): The value below which 75% of the data lies.

Interquartile Range (IQR): Q3 - Q1, the "height" of the box.

Whiskers: Extend to the smallest and largest values within 1.5 × IQR.

Outliers: Data points beyond whiskers, shown as dots or circles.



```python
data = [1,7, 8, 5, 6, 9, 12, 15, 14, 10,50]

plt.boxplot(data,
            notch=True, # creates a notched box
            vert=True, # If True, creates a Vertical Box else creates a horizontal Box
            patch_artist=True, # Box are filled with colors
            showmeans=True, # Shows the median bar pointer
            boxprops=dict(facecolor='lightyellow', color='blue'), # Determines which color is used by the boxes, faceyellow is for box color, color is for border
            medianprops=dict(color='red', linewidth=2), # SHows the median bar with color red and width 2 inches
            meanprops=dict(marker='D', markerfacecolor='green', markersize=7), # Shows the marker pointing to medianprops
            whiskerprops=dict(color='black', linewidth=1.5), # Defines the line from q1 to small value and q3 to large value
            capprops=dict(color='purple', linewidth=1.5), # Defines the external lines of the whister plot
            flierprops=dict(marker='o', color='orange', markersize=6) # Defines the marker of the outliers
           )

plt.title("Customized Box Plot")
plt.show()

```

Heatmaps

What it is:
A 2D representation of data where values are represented as colors.

When to use:
To visualize intensity or patterns in matrices.

```python
import numpy as np
data = np.random.rand(5, 5) # Creates a 5x5 array of random floating-point numbers between 0 and 1.
plt.imshow(data, cmap='gray', interpolation='nearest') # lt.imshow() displays data as an image (2D array). cmap = hot, Reds, winter, Greys. interpolation = hanning, hamming, nearest, bicubic, quadric
# Colormap to map data values to colors. 'hot' goes from black → red → yellow → white (heat colors).
# How to render pixels between data points. 'nearest' means no smoothing, pixels are shown as blocks.
plt.colorbar()  # Add color scale
plt.title('Heatmap Example')
plt.show()
```

```python
import matplotlib.pyplot as plt

# X-axis categories
quarters = ['Q1', 'Q2', 'Q3', 'Q4']

# Heights of bars for Product A
product_a = [20, 35, 30, 25]

# Heights of bars for Product B
product_b = [15, 25, 35, 30]

# Plot the first set of bars
plt.bar(
    quarters,          # X-axis categories
    product_a,         # Heights of bars
    label='Product A', # Label for legend
    color='skyblue',   # Bar color (optional)
    width=0.5,         # Width of bars (optional)
    edgecolor='black'  # Edge color of bars (optional)
)

# Plot the second set of bars stacked on top of Product A
plt.bar(
    quarters,
    product_b,
    bottom=product_a,   # Stack on top of product_a
    label='Product B',
    color='orange',
    width=0.5,
    edgecolor='black',
    alpha = 0.5
)

# Add title of the chart
plt.title("Quarterly Sales", fontsize=14)

# Label for Y-axis
plt.ylabel("Sales", fontsize=12)

# Display legend
plt.legend()

# Show the plot
plt.show()
```

| Attribute    | Description                                       |
| ------------ | ------------------------------------------------- |
| `x`          | X-axis positions or categories                    |
| `height`     | Heights of the bars                               |
| `width`      | Width of the bars (default: 0.8)                  |
| `bottom`     | Y-value to start the bar from (used for stacking) |
| `align`      | Alignment of bars: `'center'` or `'edge'`         |
| `color`      | Fill color of bars                                |
| `edgecolor`  | Color of bar edges                                |
| `label`      | Label for legend                                  |
| `hatch`      | Pattern filling (`'/', '\', 'x', 'o', etc.`)      |
| `alpha`      | Transparency of bars (0 to 1)                     |
| `tick_label` | Labels for the X-axis ticks                       |


```python
import matplotlib.pyplot as plt
import numpy as np

# Create 100 points evenly spaced between 0 and 2*pi (full circle)
theta = np.linspace(0, 2*np.pi, 100)

# Define the radius for each angle (r = sin(theta))
r = np.sin(theta)

# Create a polar plot
plt.polar(
    theta,  # Angles (in radians)
    r,      # Radius values corresponding to each angle
    color='red',      # Color of the line (optional)
    linestyle='-',     # Line style: '-', '--', '-.', ':' (optional)
    linewidth=2,       # Width of the line (optional)
    marker='o',        # Marker style for points (optional)
    markersize=4,      # Size of the markers (optional)
    label='r = sin(theta)'  # Label for legend (optional)
)

# Add a title to the plot
plt.title("Polar Plot", fontsize=14)

# Add legend if labels are provided
plt.legend()

# Show the plot
plt.show()
```

| Attribute         | Description                               |
| ----------------- | ----------------------------------------- |
| `theta`           | Angles in radians                         |
| `r`               | Radius values corresponding to each angle |
| `color`           | Line color                                |
| `linestyle`       | Line style: `'-', '--', '-.', ':'`        |
| `linewidth`       | Width of the line                         |
| `marker`          | Marker style: `'o', '^', '*', 's', etc.`  |
| `markersize`      | Size of markers                           |
| `label`           | Label for legend                          |
| `alpha`           | Transparency (0 to 1)                     |
| `markerfacecolor` | Fill color of markers                     |
| `markeredgecolor` | Edge color of markers                     |
