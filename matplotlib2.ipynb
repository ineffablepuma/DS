```python
import matplotlib.pyplot as plt

labels = ['Samsung','Apple','xiomi','others']
size = [30,25,20,25]

plt.pie(size,labels=labels, autopct='%1.0f%%',startangle=90)
plt.title("Smartphone market share")
plt.axis('equal')

plt.show()
```

```python
# pie chart chowing percentage distribution in 5 diff streams
import matplotlib.pyplot as plt

streams = ['Science','Commerce','Arts','Computer Science','Others']
students = ['120','80','100','60','40']
explode = [0.1,0,0.2,0,0] # slice lai nikalni 

plt.figure(figsize=(8,8))  #size jasto
plt.pie(students,labels=streams,autopct='%1.1f%%',startangle=90,explode=explode)
plt.title('Distribution of students in differenet streams')
plt.show()
```

```python
#histogram
# range called bins or intervals, looks like bar chart

import matplotlib.pyplot as plt

test_score = [35,54,65,80,46,97,64,87]

plt.hist(x = test_score,bins=6,edgecolor = 'black')

plt.title('Distribution of students test score(custome bins)')
plt.xlabel('Score')
plt.ylabel('Frequency')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'Height' : [150,160,170,180,190,175,165,155],
    'Weight' : [50,60,65,80,90,75,62,55],
    'Age'  : [20,21,22,23,24,25,26,27]
}

df  = pd.DataFrame(data)
print(df)

plt.figure(figsize=(8,6))
plt.scatter(df['Height'],df['Weight'],df['Age'],color='blue',marker='o')
fig = plt.figure()
```
