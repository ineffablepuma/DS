```python
# b_miles_walked_double_bar.py
import matplotlib.pyplot as plt
import numpy as np

# Data
days = ["Day 1", "Day 2", "Day 3", "Day 4", "Day 5"]
school_miles = [1.3, 1.9, 2.7, 1.6, 2.5]
vacation_miles = [0.6, 1.2, 0.9, 1.4, 2.0]

# X-axis positions
x = np.arange(len(days))
width = 0.35  # width of each bar

# Plot
plt.figure(figsize=(8,5))
plt.bar(x - width/2 , school_miles, width, label="School Week", color="skyblue")
plt.bar(x + width/2, vacation_miles, width, label="Vacation", color="salmon")

# Labels & styling
plt.xlabel("Days")
plt.ylabel("Miles Walked")
plt.title("Miles Walked: School Week vs Vacation")
plt.xticks(x, days)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
categories = ['Books & Supplies', 'Snacks & Pizza', 'Dorm Accessories', 
              'Entertainment', 'Parking Fine', 'Unspent']
amounts = [850, 65, 210, 46, 40, 39]  # numeric values
explode = [0, 0, 0, 0, 0, 0.1]       # emphasize books and unspent

# Plot
plt.figure(figsize=(8,8))
plt.pie(amounts, labels=categories, autopct='%1.1f%%', startangle=180, explode=explode)
plt.title("Gahiji's College Spending")
plt.show()
```

```python
import matplotlib.pyplot as plt

# Keep all x-axis labels
years = ["5 yrs ago", "4 yrs ago", "3 yrs ago", "2 yrs ago", "1 yr ago", "Now"]

# Remove missing data, just connect 2 yrs ago to Now
farms = [120, 200, 250, 350, 560]  # only real data

# To align data with labels, we can use indices
x = [0, 1, 2, 3, 5]  # skip index 4 ("1 yr ago")
plt.figure(figsize=(8,5))
plt.plot(x, farms, marker="o", linewidth=2, color="green")

# Set custom x-axis labels
plt.xticks(range(len(years)), years)
plt.title("Growth of Fish Farms in State")
plt.xlabel("Time")
plt.ylabel("Number of Farms")
plt.show()
plt.show()
```

```python
import matplotlib.pyplot as plt
import datetime

# Define the tasks and their start/end dates
tasks = [
    ("Review of literature", datetime.date(2025, 7, 1), datetime.date(2025, 7, 31)),
    ("Draft written", datetime.date(2025, 8, 1), datetime.date(2025, 8, 20)),
    ("Draft sent to editor", datetime.date(2025, 8, 21), datetime.date(2025, 8, 25)),
    ("Draft sent to reviewers", datetime.date(2025, 8, 26), datetime.date(2025, 9, 10)),
    ("Draft revised", datetime.date(2025, 9, 11), datetime.date(2025, 9, 30)),
    ("Article resubmitted", datetime.date(2025, 10, 1), datetime.date(2025, 10, 5)),
    ("Article published", datetime.date(2025, 12, 1), datetime.date(2025, 12, 1)),
]

# Plotting
fig, ax = plt.subplots(figsize=(10, 5))

# Create horizontal bars
for i, (task, start, end) in enumerate(tasks):
    ax.barh(i, (end - start).days + 1, left=start, height=0.5, color='skyblue')

# Format y-axis
ax.set_yticks(range(len(tasks)))
ax.set_yticklabels([task[0] for task in tasks])
ax.invert_yaxis()  # Tasks from top to bottom

# Format x-axis as dates
ax.xaxis_date()
plt.xlabel('Date')
plt.title("Gantt Chart: Zach's Article Timeline")
plt.tight_layout()
plt.show()
```
