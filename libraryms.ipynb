```python
class Book:
    def __init__(self, title, author, published_date):
        self.title = title
        self.author = author
        self.published_date = published_date
        
    def __str__(self):
        return f"{self.title} by {self.author}" 
    
# DRY - Don't Repeat Yourself
class eBook(Book):
    def __init__(self, title, author, published_date, file_size):
        super().__init__(title, author, published_date)
        self.file_size = file_size
        
    def __str__(self):
        return f"{super().__str__()} - eBook ({self.file_size})"
        
class Library:
    def __init__(self):
        self.books = []   # store actual book objects, not just titles

    def add_book(self, book):
        for b in self.books:
            if b.title == book.title:   # check duplicate by title
                print("Book already exists.")
                return
        self.books.append(book)
        print(f"‚úÖ {book.title} added to library.")

    def remove_book(self, title):
        for b in self.books:
            if b.title == title:
                self.books.remove(b)
                print(f"‚ùå {title} removed from library.")
                return
        print("Book not found.")
        
    def search_book(self, title):
        for i, b in enumerate(self.books):
            if b.title == title:
                print(f"üîé Found: {b} at index {i}")
                return
        print("Book not found.")
            
    def view_books(self):
        if not self.books:
            print("üìö No books in library.")
        else:
            print("\nüìö Library Collection:")
            for i, book in enumerate(self.books, 1):
                print(f"{i}. {book}")
            print()

# ------------------ MAIN LOOP ------------------ #
library = Library()

while True:
    print("\n--- Library Menu ---")
    print("1. Add Book")
    print("2. Add eBook")
    print("3. View Books")
    print("4. Search Book")
    print("5. Remove Book")
    print("6. Exit")
    
    choice = input("Enter your choice: ")
    
    if choice == "1":
        title = input("Enter book title: ")
        author = input("Enter author name: ")
        published_date = input("Enter published date: ")
        book = Book(title, author, published_date)
        library.add_book(book)

    elif choice == "2":
        title = input("Enter eBook title: ")
        author = input("Enter author name: ")
        published_date = input("Enter published date: ")
        file_size = input("Enter file size (e.g., 5MB): ")
        ebook = eBook(title, author, published_date, file_size)
        library.add_book(ebook)

    elif choice == "3":
        library.view_books()

    elif choice == "4":
        title = input("Enter book title to search: ")
        library.search_book(title)

    elif choice == "5":
        title = input("Enter book title to remove: ")
        library.remove_book(title)

    elif choice == "6":
        print("üëã Exiting program. Goodbye!")
        break

    else:
        print("‚ö†Ô∏è Invalid choice. Please try again.")
```
