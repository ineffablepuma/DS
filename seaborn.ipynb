```python
import numpy as np
import seaborn as sns

sns.set(style="white")  #background white
r = np.random.RandomState(10)  #seed for reproducibility
d = r.normal(size = 100)  #100 random numbers from a normal distribution with mean = 0 and sd= 1

sns.histplot(d,kde = True,color="m")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
fmri = sns.load_dataset('fmri')
fmri
#plt.scatter(fmri['timepoint'],fmri['signal'])
#sns.scatterplot(x='timepoint', y='signal', data=fmri)
sns.barplot(x='event', y='signal', data=fmri, estimator='mean')
plt.show()
```

```python
fmri
```

```python
region_counts = fmri['region'].value_counts()
print(region_counts)
plt.pie(region_counts, labels=region_counts.index, autopct='%1.1f%%', startangle=140)
plt.show()
```

```python
plt.hist(fmri['signal'], bins=30, color='skyblue', edgecolor='black')
plt.show()
```

```python
sns.histplot(data=fmri, x='signal', bins=30, kde=True)
plt.show()
```

```python
sns.boxplot(x='event', y='signal', data=fmri)
plt.show()
```

```python
sns.relplot(x="timepoint",y="signal",data = fmri,kind = 'line')
plt.show()
```

```python
sns.relplot(x="timepoint",y="signal",data = fmri,kind = 'line',hue='event')
plt.show()
```

```python
sns.relplot(x="timepoint",y="signal",data = fmri,kind = 'line',hue='event',col='region')
plt.show()
```

```python
sns.relplot(x="timepoint",y="signal",data = fmri,hue='event',col='region')
plt.show()
```

```python
sns.relplot(x="timepoint",y="signal",data = fmri,kind = 'violin',hue='event',col='region')
plt.show()
```

```python
sns.catplot(x='event',y='timepoint',data=fmri,kind='violin')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

fmri = sns.load_dataset('fmri')

sns.regplot(x='timepoint', y='signal', data=fmri, scatter_kws={'alpha':0.1})
plt.title('Regression plot of Signal vs Timepoint')
plt.show()
```
