```python
#line plot
import matplotlib.pyplot as plt
x = (1,3,4,6,7)
y = (2,3,5,6,8)

plt.plot(x,y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Sample Line Plot')
plt.show()
```

```python
import matplotlib.pyplot as plt
x = (1,3,4,6,7)
y = (2,3,5,6,8)

plt.plot(x,y,marker='o',color='green')  #marker dherai hunxan
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Sample Line Plot')
plt.show()
```

```python
import matplotlib.pyplot as plt
x = (1,3,4,6,7)
y = (2,3,5,6,8)

plt.scatter(x,y,marker='o',color='green')  #scatter
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
category = ['A','B','C','D']
values = [3,34,65,78]
plt.bar(category,values,color='green')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
y = np.array([34,56,87,45])
plt.pie(y)
plt.show()
```
